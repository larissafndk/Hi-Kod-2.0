```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1
df = pd.read_csv('StudentsPerformance.csv')  

print(df.head())
```

```python
#2
#kadın-erkek sayısına bakalım

cinsiyet_sayilari = df['gender'].value_counts()

# Sonuçları yazdıralım
print(cinsiyet_sayilari)



```

```python
#3
#kadın-erkek sayısına histogram grafiği ile bakalım
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', palette='viridis')
plt.title('Kadın ve Erkek Sayısı')
plt.xlabel('Cinsiyet')
plt.ylabel('Sayı')
plt.show()
```

```python
#4
#race/ethnicity sütununda kaç farklı grup olduğuna bakalım
benzersiz= df['race/ethnicity'].unique()
grup_sayisi = df['race/ethnicity'].nunique()
print("Farklı gruplar:", benzersiz)
print("Farklı grup sayısı:", grup_sayisi)
```

```python
#5
#yukarıda bulduğumuzu görselleştirelim
grup_sayilari = df['race/ethnicity'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=grup_sayilari.index, y=grup_sayilari.values, palette='viridis')
plt.title('Race/Ethnicity Gruplarının Dağılımı')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Sayı')
plt.xticks(rotation=45)
plt.show()
```

```python
lunch_Sayi = df["test preparation course"].value_counts()
print(lunch_Sayi)
```

```python
#9
#gender sütunundaki değerler için ortalama math score, reading score, writing score değerlerini bulalım.
notOrt = df[~df['math score'].apply(lambda x: str(x).isnumeric())]
print(notOrt)
```

```python
#10
#lunch sütunundaki değerler için ortalama math score, reading score, writing score değerlerini bulalım.
lunchNot = df.groupby('lunch')[['math score', 'reading score', 'writing score']].sum()
ogrenciSayi = df['lunch'].value_counts()

ortalama_notlar = lunchNot.div(ogrenciSayi, axis=0)
print(ortalama_notlar)
```

```python
#11
#test preparation coursesütunundaki değerler için ortalama math score, reading score, writing score değerlerini bulalım.
test_notlari = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].sum()
ogrenci_sayisi = df['test preparation course'].value_counts()
ortalama_notlar = test_notlari.div(ogrenci_sayisi, axis=0)
print(ortalama_notlar)
```
