```python
def yas_hesap(dogum_yili):
    from datetime import datetime
    simdi= datetime.now()
    mevcut_yil = simdi.year
    yas = mevcut_yil - dogum_yili
    return yas
def emeklilik_durum(dogum_yili, isim):
    yas = yas_hesap(dogum_yili)
    emeklilik = 65
    if yas >= emeklilik:
        print("Emekli oldunuz")
    else:
        kalan_yil = emeklilik - yas
        print("{} emekliliğine {} yıl kaldı.".format(isim, kalan_yil))
dogum_yili = int(input("Doğum yılınızı girin: "))
isim = input("İsminizi girin: ")
emeklilik_durum(dogum_yili, isim)










```
